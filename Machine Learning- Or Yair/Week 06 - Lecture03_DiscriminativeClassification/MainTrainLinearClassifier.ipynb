```python
import numpy             as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.size'] = 16
np.random.seed(1)
```

### Load moons data:

```python
from sklearn.datasets import make_moons

N      = 500
mX, vY = make_moons(n_samples=N, noise=.1)

mX.shape, vY.shape
```

```python
vY[vY == 0] = -1
vY
```

#### Plot the data:

```python
plt.figure (figsize=(5, 5))
plt.scatter(mX[vY==-1,0], mX[vY==-1,1], color='r', s=100, edgecolor='k')
plt.scatter(mX[vY== 1,0], mX[vY== 1,1], color='b', s=100, edgecolor='k')
plt.xlabel ('$x_1$')
plt.ylabel ('$x_2$')
plt.axis   ('equal')
plt.grid   (True)
plt.show   ()
```

#### Define:
$$\tilde{\boldsymbol{X}}=\left[\begin{matrix}-1 & - & x_{1} & -\\
-1 & - & x_{2} & -\\
 & \vdots\\
-1 & - & x_{N} & -
\end{matrix}\right]\in\mathbb{R}^{N\times3}$$

```python
mTildeX = np.concatenate([-np.ones((N, 1)), mX], axis=1)
mTildeX.shape, mTildeX
```

#### Sigmoid:
$$\sigma\left(\alpha\right)=2\frac{\exp\left(\alpha\right)}{1+\exp\left(\alpha\right)}-1$$
#### Sigmoid derivative:
$$\frac{\text{d}\sigma\left(\alpha\right)}{\text{d}\alpha}=2\frac{\exp\left(\alpha\right)}{\left(1+\exp\left(\alpha\right)\right)^{2}}$$
#### Loss:
$$\arg\min_{\tilde{\boldsymbol{w}}}\tilde{J}\left(\tilde{\boldsymbol{w}}\right)=\arg\min_{\tilde{\boldsymbol{w}}}\frac{1}{4N}\left\Vert \sigma\left(\tilde{\boldsymbol{X}}\tilde{\boldsymbol{w}}\right)-\boldsymbol{y}\right\Vert _{2}^{2}$$
#### Loss gradient:
$$\nabla_{\tilde{\boldsymbol{w}}}\tilde{J}\left(\tilde{\boldsymbol{w}}\right)=\frac{1}{2N}\tilde{\boldsymbol{X}}^{T}\sigma'\left(\text{diag}\left(\tilde{\boldsymbol{X}}\tilde{\boldsymbol{w}}\right)\right)\left(\sigma\left(\tilde{\boldsymbol{X}}\tilde{\boldsymbol{w}}\right)-\boldsymbol{y}\right)$$

```python
def sigmoid(α):
    return 2 * np.exp(α) / (np.exp(α) + 1) - 1

def Dsigmoid(α):
    return 2 * np.exp(α) / (np.exp(α) + 1)**2

def Loss(vTildeW):
    return np.linalg.norm(sigmoid(mTildeX @ vTildeW) - vY[:,None])**2 / (4*N)

def ComputeGradient(vTildeW):
    return (mTildeX.T * Dsigmoid(mTildeX @ vTildeW).T) @ (sigmoid(mTildeX @ vTildeW) - vY[:,None]) / (2*N)
```

__Plot funciton:__

```python
def PlotW(vTildeW, fig):
    b     = vTildeW[0] .squeeze()
    vW    = vTildeW[1:].squeeze()
    vIdx1 = (mX @ vW - b) < 0
    vHatY = np.sign(mX @ vW - b)
    
    vLoss .append(Loss(vTildeW))        #-- global
    vError.append(np.mean(vHatY != vY)) #-- global
    
    ax1 = fig.axes[0]
    ax2 = fig.axes[1]
    if fig.axes:
        ax1.cla()
        ax2.cla()

    v = np.array([-3, 3])
    ax1.scatter(mX[vIdx1, 0], mX[vIdx1, 1], color='r',      s=100, edgecolor='k')
    ax1.scatter(mX[~vIdx1,0], mX[~vIdx1,1], color='b',      s=100, edgecolor='k')
    ax1.plot   (v, (b - vW[0] * v) / vW[1], color='k',      lw=2)
    ax1.arrow  (0, 0, vW[0], vW[1],         color='orange', width=.05)
    
    ax1.grid      ()
    ax1.axvline   (x=0, c='k')
    ax1.axhline   (y=0, c='k')
    ax1.axis      ('equal')
    ax1.axis      ([-1.9, 2.5, -2.2, 2.2])
    ax1.set_title (r'$f_{{w},b}\left({x}\right)={sign}\left({w}^{T}{x}-b\right)$')
    ax1.set_xlabel('$x_1$')
    ax1.set_ylabel('$x_2$')
    
    ax2.plot      (vError, 'k', lw=2, label=r'${J}\left(\tilde{{w}}\right)$')
    ax2.plot      (vLoss,  'm', lw=2, label=r'$\tilde{J}\left(\tilde{{w}}\right)$')
    ax2.set_xlabel('iterations')
    ax2.grid      ()
    ax2.legend    ()
        
    fig.canvas.draw()
    plt.pause(1e-6)
```

#### Apply gradient descent:
$$\tilde{\boldsymbol{w}}_{k+1}=\tilde{\boldsymbol{w}}_{k}-\mu\nabla_{\tilde{\boldsymbol{w}}}\tilde{J}\left(\tilde{\boldsymbol{w}}_{k}\right)$$
(Not supported on Colab)

```python
%matplotlib notebook

vLoss  = []
vError = []

#-- Some init vector w:
vTildeW = np.random.randn(3, 1)
# vTildeW = np.zeros((3, 1))
vTildeW = np.array([[0,-2.0,2]]).T

#-- Apply gradient descent:
K       = 200
µ       = .3
fig, _  = plt.subplots(1, 2, figsize=(12, 5))
for ii in range(K):
    #-- Gradient descent step:
    vTildeW -= µ * ComputeGradient(vTildeW)
    
    #-- Plot:
    PlotW(vTildeW, fig)    
    
plt.show()
```
