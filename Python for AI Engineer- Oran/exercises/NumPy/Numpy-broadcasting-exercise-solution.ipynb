```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
n = 400
a = np.linspace(0, 1, n)
b = np.ones(n)
```

#### Drill #1

```python
m = a * b[:,np.newaxis]
# or:
# m = np.multiply.outer(b, a)
plt.imshow(m, 'gray')
plt.colorbar();
```

#### Drill #2

```python
m = a[:,np.newaxis] * b
# or:
# m = np.multiply.outer(a, b)
plt.imshow(m, 'gray')
plt.colorbar();
```

#### Drill #3

```python
m = a/2 + a[:,np.newaxis]/2
# or:
# m = np.add.outer(a/2, a/2)
plt.imshow(m, 'gray')
plt.colorbar();
```

#### Drill #4

```python
m = a / 2 + a[:,np.newaxis]/ 2 >= 0.5
# or:
# m = np.add.outer(a/2, a/2) >= 0.5
plt.imshow(m, 'gray');
```
