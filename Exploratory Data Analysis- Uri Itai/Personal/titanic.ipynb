## Let's check the skew

```python
sns.displot(train["Age"], color="b", label="Skewness : %.2f"%(train["Age"].skew())), sns.displot(train["Fare"], color="b", label="Skewness : %.2f"%(train["Fare"].skew()))
```

- Age distribution is normal.
- Fare distribution is skewed. Will log

```python
train['FareLog'] = train.Fare.apply(lambda x: np.log(x) if x > 0 else 0)
sns.displot(train["Age"], color="b", label="Skewness : %.2f"%(train["Age"].skew())), sns.displot(train["FareLog"], color="b", label="Skewness : %.2f"%(train["FareLog"].skew()))
```

- It shows even though there were less young people, more survived. This is probably contributed by the fact babies were saved.
- Fare distibution shows most paid average fair, but the higher fare people survived. Could be that those who paid more were closer to exits.
