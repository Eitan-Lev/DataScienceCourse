```python
def outlier1(arr):
    return [x for x in arr if np.abs(x - np.mean(arr))/np.std(arr) > 3]

def outlier2(arr):
    mad_arr = np.median([np.abs(x - np.median(arr)) for x in arr])
    return [x for x in arr if np.abs(x - np.mean(arr))/mad_arr > 3]

def outlier3(arr):
    q75 = np.quantile(arr, 0.75)
    q25 = np.quantile(arr, 0.25)
    upper = 2*q75 - q25
    lower = q75 - 2 * q25
    return [x for x in arr if x < lower or x > upper]

def all_outliers(arr):
    return outlier1(arr) + outlier2(arr) + outlier3(arr)

def remove_outlier1(arr):
    return [x for x in arr if np.abs(x - np.mean(arr))/np.std(arr) <= 3]
def remove_outlier1(df):
    return [x for x in arr if np.abs(x - np.mean(arr))/np.std(arr) <= 3]

def remove_outlier2(arr):
    mad_arr = np.median([np.abs(x - np.median(arr)) for x in arr])
    return [x for x in arr if np.abs(x - np.mean(arr))/mad_arr <= 3]

def remove_outlier3(arr):
    q75 = np.quantile(arr, 0.75)
    q25 = np.quantile(arr, 0.25)
    upper = 2*q75 - q25
    lower = q75 - 2 * q25
    return [x for x in arr if x < lower or x <= upper]

def remove_all_outliers(arr):
    return outlier1(arr) + outlier2(arr) + outlier3(arr)

def get_limit_by_quantiles(arr):
    q75 = np.quantile(arr, 0.75)
    q25 = np.quantile(arr, 0.25)
    upper = 2*q75 - q25
    lower = q75 - 2 * q25
    return upper, lower
```

Removing outliers

```python
z_scores = stats.zscore(df)
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df = df[filtered_entries]
df
```

removing duplicates

```python
df.drop_duplicates()
```

```python
df2 = df[['BMI','Age']]
df2 = df2.sort_values('Age')
# plt.plot(df2.Age, df2.BMI)
plt.scatter(df2.Age, df2.BMI)
plt.title('BMI by age')
plt.show()
df3 = df[['Outcome','Age']]
df3 = df3.sort_values('Age')
# plt.plot(df3.Age, df3.Outcome)
# plt.scatter(df3.Age, df3.Outcome)
plt.bar(df3.Outcome, df3.Age)
plt.title('Outcome by age')
plt.show()
df4 = df[['Outcome','BMI']]
df4 = df4.sort_values('BMI')
plt.bar(df4.Outcome, df4.BMI)
plt.title('Outcome by BMI')
plt.show()
```
