```python
import math as math
from sympy import *
```

```python
x = Symbol('x')
y = Symbol('y')
```

```python
def grad(f):
    return f.diff(x) + f.diff(y)
def grad(f, x, y):
    return grad(f)
```

```python
f = x + y**3
```

```python
grad(f)
```

```python
import autograd.numpy as np
from autograd import grad

def function(x):
    return 0.5 * np.dot(np.dot(np.transpose(x), A), x) + np.dot(np.transpose(b), x)

A = A = np.zeros((2, 2))
n = A.shape[0]
A[range(n), range(n)] = 1
b = np.ones(2)
x = np.ones(2)
grad(function)(x)
```

```python
def f(x,y):
    return x + y**3
def derivative_f(x,y,h):
    x_deriv, y_deriv = ((f(x+h,y)-f(x-h,y))/(2*h), (f(x,y+h)-f(x,y-h))/(2*h))
    return x_deriv, y_deriv
```

```python
derivative_f(1,5,1e-3)
```
