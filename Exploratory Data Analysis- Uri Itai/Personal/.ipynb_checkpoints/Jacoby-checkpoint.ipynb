```python
import numpy as np
from autograd import grad, jacobian
```

```python
A = np.array([[3,1,1],
              [1,2,0],
              [0,0,1]])


b = np.array([[2,1,3]]).T
```

```python
U = np.triu(A)
L = np.tril(A)
# D = np.diag(np.diag(A))
D = np.tril(A)-np.tril(A,-1)
U,D,L
```

```python
x = np.matrix([[1],[1],[1]])
invD = np.linalg.inv(D)
for i in range(50):
    x = np.matmul(invD, b-np.matmul(L+U,x))

x, np.matmul(A,x)
```
