```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/to_run.csv')
```

```python
df.head()
```

```python
res = []
for i,v in enumerate(df.appear.tolist()):
    if i == 0:
        res.append(1)
        continue
    if v == True:
        res.append(0)
        continue
    res.append(res[i-1]+1)
    
df['run_length'] = res
```
